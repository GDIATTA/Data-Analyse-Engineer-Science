## TP Matplotlib

Je travaillerai avec l'ensemble de données de vente d'un détaillant en ligne. Les données sont collectées sur une période de trois ans : 2012 à 2015. Elles contiennent les informations sur les ventes réalisées par l'entreprise.

Les produits capturés appartiennent à trois catégories :

     Meubles
     Fournitures de bureau
     Technologie


```python
# importer numpy et  matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creer un array avec des categories de produits 
product_category = np.array(['Furniture', 'Tech', 'Office_Supply'])
sales = np.array([4110451.90, 4744557.50, 3787492.52])
```

## 1- Bar Graph: Tracer les ventes pour
chaque catégorie de produits

- Un graphique à barres utilise des barres pour afficher des comparaisons entre des catégories de données.

- Un axe aura généralement des valeurs numériques ou des mesures,
- L'autre décrira les types de catégories comparées ou les dimensions


```python
# plotting the bar graph with product categories on x-axis and sales amount of y-axis
plt.bar(product_category, sales,width=0.8)

# adding title to the graph
plt.title("Sales across Product Categories", fontsize= 20, fontweight= 5, color = 'Green')

# labeling axes
plt.xlabel("Product_Categories", fontsize= 20, fontweight= 5, color = 'Green') 
plt.ylabel("Sales", fontsize= 20, fontweight= 5, color = 'Green')

# Modifying the ticks to show information
#plt.xticks(rotation ='vertical')
plt.xticks(rotation =45)
plt.yticks(rotation='horizontal')
# necessary command to display the created graph
plt.show()
```


## 2- Scatter Chart - Tracer les ventes en fonction des profits 
Scatter plots (nuages de points) sont utilisés lorsque vous souhaitez montrer la relation entre deux faits ou mesures.

- Le profit est mesuré dans chaque catégorie de produits

```python
product_category = np.array(['Furniture', 'Tech', 'Office_Supply','Furniture', 'Tech', 'Office_Supply'])
sales = np.array([4110451.90, 4744557.50, 3787492.52,2980451.90, 3744557.50, 2787492.52])
profit = np.array([1213.46, 1814.13, 1485.7,2213.46, 2814.13, 2485.7])
```

```python
# plotting scatter chart

plt.scatter(profit, sales )#alpha= .7, s = 50

# Adding and formatting title
plt.title("Sales versus Profits of product Categories\n",fontsize= 20, fontweight= 5, color = 'Green')

# Labeling Axes
plt.xlabel("Profit", fontsize= 20, fontweight= 5, color = 'Brown')
plt.ylabel("Sales", fontsize= 20, fontweight= 5, color = 'Brown')

plt.show()
```

## 3- Scatter Chart - Tracer les ventes en fonction des profits de maniere plus sophistiquee

```python
plt.scatter(profit[product_category == "Furniture"], sales[product_category == "Furniture"], 
            c= 'Green', s = 150, label="Furniture")

plt.scatter(profit[product_category == "Tech"], sales[product_category == "Tech"], 
            c= 'Yellow', s = 100, label="Tech" )

plt.scatter(profit[product_category == "Office_Supply"], sales[product_category == "Office_Supply"], 
            c= 'Cyan',  s = 50, label="Office_Supply" )

# Adding and formatting title
plt.title("Sales versus Profits across Product Categories\n", fontsize= 20, fontweight= 5, color = 'Green')

# Labeling Axes
plt.xlabel("Profit",fontsize= 15, fontweight= 5, color = 'Brown')
plt.ylabel("Sales", fontsize= 15, fontweight= 5, color = 'Brown')

plt.legend()

plt.show()
```

# 4- Line Chart - Plot Ventes en 2015

A line chart ou un graphique linéaire ou un graphique linéaire ou un graphique à courbes est un type de graphique qui affiche des informations sous la forme d'une série de points de données appelés « marqueurs » reliés par des segments de ligne droite. Plus couramment utilisé avec les données temporelles.

```python
import numpy as np
import matplotlib.pyplot as plt

months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
sales = np.array([241268.56, 184837.36, 263100.77, 242771.86, 288401.05, 401814.06, 258705.68, 456619.94, 481157.24, 422766.63, 555279.03, 503143.69])
```

```python
plt.plot(months, sales,marker="*")

# Adding and formatting title
plt.title("Sales across 2015\n", fontsize= 20, fontweight= 5, color = 'Green')

# Labeling Axes
plt.xlabel("Months", fontsize= 12, fontweight= 5, color = 'Brown')
plt.ylabel("Sales", fontsize= 12, fontweight= 5, color = 'Brown' )



plt.xticks(rotation=45)

plt.show()
```

# 5- Histogram

Un histogramme est un tableau de fréquence qui enregistre l'occurrence d'une entrée ou d'un élément dans un ensemble de données. Il est utile lorsque vous souhaitez comprendre la distribution d'une série donnée.

```python
import numpy as np
import matplotlib.pyplot as plt

age=np.array([21,23,22,24,24,23,23,22,23,24])
#plotting a hitogram
plt.hist(age,bins=10,edgecolor="blue",color="yellowgreen")
plt.show()
```

# 7- SubPlots - Tracer les Ventes pour divers marchés pour les années 2012 à 2015

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.array(['2012', '2013', '2014', '2015'])

sales_africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])

sales_USCA = np.array([492756.60, 486629.30, 627634.98, 757108.13])

sales_LATAM = np.array([385098.15, 464733.29, 608140.77, 706632.93])

sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])

sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
```

```python
fig, ax = plt.subplots(ncols=4)

europe, = ax[0].plot(years, sales_Europe)
europe.set_label("Europe")
europe.set_color("red")
ax[0].set_title('Sales in Europe')

usca = ax[1].bar(years, sales_USCA)
usca.set_label("USCA")
ax[1].set_title('Sales in USCA')

africa = ax[2].scatter(years, sales_africa)
africa.set_label("Africa")
ax[2].set_title('Sales in Africa')

asia = ax[3].bar(years, sales_Asia_Pacific, width = 0.5, color='royalblue', label ='Asia Pacific')

latam = ax[3].bar(years, sales_LATAM, width = 0.5,color='seagreen', bottom = sales_Asia_Pacific, label = 'LATAM')

ax[3].set_title('Sales in Asia Pacific and LATAM')
ax[3].legend()

fig.set_size_inches(20.5, 5.5, forward=True)

plt.show()
```
