Calculer l'âge moyen, mimimum, maximum et rénommez les colonne

```python
from pyspark.sql.functions import mean, min, max

df.agg(
    mean("Age").alias("mean_age"), 
    min("Age").alias("min_age"),
    max("Age").alias("max_age")
).show()
```

Affichez l'âge moyen des passagers groupés par classe et port de départ et triez la sortie par classe et port de départ
